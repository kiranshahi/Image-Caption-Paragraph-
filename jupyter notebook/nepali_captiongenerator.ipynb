```python


# Set up the translator
translator = Translator(service_urls=['translate.google.com'])

# Set the input and output file paths
input_file = '/content/drive/MyDrive/ml/caption/caption_text.txt'
output_file = '/content/drive/MyDrive/captions_nepali.txt'

# Function to translate a caption from English to Nepali
def translate_caption(caption):
    translation = translator.translate(caption, src='en', dest='ne')
    return translation.text

# Read the input file and translate the captions
with open(input_file, 'r') as f:
    lines = f.readlines()

translated_lines = []
for line in lines:
    if line.strip() != '':
        line = line.strip()
        image_caption = line.split('   ')
        if len(image_caption) >= 2:
            image_name = image_caption[0].strip()
            caption = image_caption[1].strip()
            translated_caption = translate_caption(caption)
            translated_line = f"{image_name}: {translated_caption}"
            translated_lines.append(translated_line)
# Save the translated captions to the output file
with open(output_file, 'w+') as f:
    f.writelines('\n'.join(translated_lines))

print('Caption translation completed.')
```
